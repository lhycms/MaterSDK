```python
from matersdk.io.publicLayer.structure import DStructure
```

# 1. 读取结构

```python
file_path = "/data/home/liuhanyu/hyliu/code/matersdk/demo/feature/movement/LiSi.config"
file_format = "pwmat"

structure = DStructure.from_file(
                file_path=file_path,
                file_format=file_format
)
print(structure)
```

# 2. 获取结构信息 + 微扰

```python
### Step 1. 获取结构信息
lattice = structure.lattice.matrix
species = structure.species
cart_coords = structure.cart_coords

print(lattice)
```

```python
### Step 2. 微扰：lattice + 0.5
lattice = lattice + 0.5
print(lattice)
```

# 3. 产生新结构

```python
new_structure = DStructure(
                    lattice=lattice,
                    species=species,
                    coords=cart_coords,
                    coords_are_cartesian=True   # 使用笛卡尔坐标，而不是分数坐标
)
```
